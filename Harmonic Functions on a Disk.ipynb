# Harmonic Functions on a Disk, with Weird Boundary Conditions

```python
#Relaxation Method

import math

#Create Empty 100x100 Grid of Zeroes

r2space = []
for i in range(0,100):
    r2space.append(100*[0])
    
iterly = [] #iterability
for i in range(0,100):
    iterly.append(100*[1])

#angle finder

def getangle(i,j):
    
    x = j-50
    y = i-50
    
    if x == 0 and y > 0:
        return math.pi/2
    elif x == 0 and y < 0:
        return 3*math.pi/2
    elif x > 0:
        return math.atan(y/x)
    elif x < 0:
        return math.atan(y/x)+math.pi
        
#Boundary Conditions

r=45

for i in range(0,100):
    for j in range(0,100):
        if math.sqrt((i-50)**2+(j-50)**2)>r:
            iterly[i][j]=0
        if round(math.sqrt((i-50)**2+(j-50)**2))==r:          
            
            r2space[i][j]=100*math.sin(3*getangle(i,j))
            iterly[i][j]=0
            
            #Some Interesting Functions to Try:
            #100*math.sin(n*getangle(i,j)) for n = 1,2,3,4,5, etc        

#Iterator

k = 1000 #Iterations Desired
n=0
while n<k:
    n=n+1
    for i in range(1,99):
        for j in range(1,99):
            if iterly[i][j]==1:   
                S = (r2space[i-1][j]+r2space[i][j-1]+r2space[i][j+1]+r2space[i+1][j])/4
                r2space[i][j] = S
                
print("Iterations Applied:",k)
```

```python
import plotly.graph_objects as go

fig = go.Figure(data =
    go.Contour(
        z=r2space,
        contours=dict(
            start=-100,
            end=100,
            size=10,
        ),
    ))
fig.show()
```
